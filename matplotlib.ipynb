# Matplotlib

`Matplotlib` 是 `Python` 中最常用的绘图库之一，广泛用于数据可视化。它提供了丰富的绘图功能，可生成折线图、柱状图、散点图、饼图、直方图等多种图形。


`pyplot` 是 `Matplotlib` 的核心模块，提供类似 MATLAB 的绘图接口。它以函数形式控制图形的创建、显示和样式设置，支持绘制折线图、柱状图、散点图等多种图表。通过简单命令即可快速生成高质量可视化结果，广泛应用于数据分析与科研绘图。通常来说，我们使用简短的别名`plt`来导入它

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

## 常见函数

| 函数名                 | 功能说明                |
| :------------------ | :------------------ |
| `plot()`            | 绘制折线图或点线图，是最常用的绘图函数 |
| `bar()`             | 绘制垂直柱状图             |
| `barh()`            | 绘制水平柱状图             |
| `scatter()`         | 绘制散点图，用于显示变量间关系     |
| `pie()`             | 绘制饼图，展示各部分占比        |
| `hist()`            | 绘制直方图，用于展示数据分布      |
| `boxplot()`         | 绘制箱线图，显示数据分布的离散程度   |
| `imshow()`          | 显示图像数据（如矩阵或图片）      |
| `title()`           | 设置图表标题              |
| `xlabel()`          | 设置 x 轴标签            |
| `ylabel()`          | 设置 y 轴标签            |
| `legend()`          | 添加图例                |
| `xlim()` / `ylim()` | 设置坐标轴显示范围           |
| `grid()`            | 显示或隐藏网格线            |
| `show()`            | 显示绘制的图形             |


## 基本操作

### 绘制点和线

````python
x1 = [0, 1, 2, 3, 4]
y1 = [0, 2, 6, 2, 8]

plt.plot(x1, y1, color='blue', linestyle='--', linewidth=2, marker='s')
plt.title('Line Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````

在上图中，我们利用到了`plot()`函数，`plot()`函数是绘制二维图形的基础，我们还可以利用`scatter`函数来绘制点，如下图所示

````python
x2 = [1, 2, 3, 4, 5]
y2 = [2, 9, 3, 7, 11]

plt.scatter(x2, y2, color='red', marker='o', label='Data Points')
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
````

从上面的两张图里也能看出，图上面的点、线是可以通过设置属性变得更加**美观**的，我们来列一个表格简单介绍一下所有的这些属性应该怎么使用🎨

| 属性                              | 参数类型 / 示例                                              | 说明                           |
| :------------------------------ | :----------------------------------------------------- | :--------------------------- |
| **color / c**                   | `'red'`, `'blue'`, `'g'`, `'#FF8800'`                  | 设置颜色，可用颜色名、简写或十六进制值          |
| **linestyle / ls**              | `'-'`, `'--'`, `':'`, `'-.'`, `'none'`                 | 设置线型（实线、虚线、点线、点划线、无线）        |
| **linewidth / lw**              | `1`, `2.5`, `5`                                        | 设置线宽                         |
| **marker**                      | `'o'`, `'s'`, `'^'`, `'x'`, `'+'`, `'d'`, `'*'`, `'.'` | 设置点的形状                       |
| **markersize / ms**             | `6`, `10`, `15`                                        | 设置点的大小                       |
| **markeredgecolor / mec**       | `'black'`, `'k'`, `'#333333'`                          | 设置点边缘颜色                      |
| **markerfacecolor / mfc**       | `'white'`, `'red'`, `'none'`                           | 设置点内部填充颜色                    |
| **alpha**                       | `0.0 ~ 1.0`                                            | 设置透明度，数值越小越透明                |
| **label**                       | `'Data1'`, `'sin(x)'`                                  | 设置图例文字，与 `plt.legend()` 配合使用 |
| **grid()**                      | `plt.grid(True)`                                       | 是否显示网格线                      |
| **xlim(), ylim()**              | `plt.xlim(0,10)`                                       | 设置坐标轴显示范围                    |
| **title(), xlabel(), ylabel()** | 文本字符串                                                  | 设置标题和坐标轴标签                   |


下面的图形是使用了诸多属性后的示例，你也可以在此基础上继续进行美观调整💪

````python
x3 = np.linspace(0, 3, 7)
y3 = 2 ** x3 - 5

plt.plot(
    x3, y3,
    color='purple',        # 线的颜色
    linestyle='--',        # 虚线
    linewidth=2,           # 线宽
    marker='o',            # 圆形标记
    markersize=8,          # 标记大小
    markerfacecolor='white',  # 标记填充色
    markeredgecolor='black',  # 标记边框色
    alpha=0.8,             # 透明度
    label='y = 2x^2-5'
)

plt.title('Line and Marker Style Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

## 绘制函数图

在`matplotlib`中，我们也可以绘制一些**函数**的曲线

````python
x4 = np.linspace(0, 2 * np.pi, 200)  # 在 0 到 2π 之间取 200 个点
y4 = np.sin(x4)

plt.plot(x4, y4, label='y = sinx')
plt.title('sinx')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````

在同一张图里，也可以放两条及更多的函数曲线

````python
x5 = np.linspace(0, 2 * np.pi, 200)  # 在 0 到 2π 之间取 200 个点
y5 = np.sin(x5)
y6 = np.cos(x5)

plt.plot(x5, y5, label='sin')
plt.plot(x5, y6, linestyle='--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
````

## 绘制柱形图

* `bar(x, height)`：绘制垂直柱形图
* `color`：柱子颜色
* `width`：柱宽
* 若需水平展示，可使用 `plt.barh()`


````python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 10]

plt.bar(categories, values, color='orange', width=0.6)
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
````

## 绘制饼状图

* `labels`：每部分标签
* `sizes`：每部分占比
* `autopct`：显示百分比格式
* `explode`：突出某部分
* `startangle`：旋转起始角度


````python
labels = ['Apple', 'Banana', 'Cherry', 'Durian']
sizes = [30, 25, 25, 20]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = [0.05, 0, 0, 0]  # 突出第一个部分

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=90
)
plt.title('Pie Chart Example')
plt.show()
````

## 绘制直方图

* `hist(data)`：绘制直方图
* `bins`：直方图的区间数
* `edgecolor`：柱子边框颜色
* `alpha`：透明度
* 常用于连续型数据的分布可视化

````python
import numpy as np

data = np.random.randn(1000)  # 生成1000个符合正态分布的随机数

plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.8)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

| 保存图像 | `plt.imsave()`               | 将 NumPy 数组保存为图像 |

### 图片的旋转

````python
im_flipped_h = np.fliplr(img)  # 水平翻转
plt.imshow(im_flipped_h)
plt.show()
````

````python
from matplotlib.image import imread
im = imread('assets/lena.png')
plt.imshow(im)
plt.show()
````

### 图片的裁剪

````python
im_cut = im[50:300, 50:300]
plt.imshow(im_cut)
plt.show()
````

### 图片的变暗

````python
im_darker = im.copy()

# 显示图像
plt.figure(figsize=(10,6))

for i in range(1,5):
    plt.subplot(2,2,i)
    x = 1 - 0.2*(i-1)
    plt.axis('off') #hide coordinate axes
    plt.title('x={:.1f}'.format(x))
    im_darker = np.clip(im_darker, 0, 1)  # 防止数值超出范围
    plt.imshow(im_darker * x)

plt.show()
````

### 图片的颜色改变

````python
# 修改RGB颜色
im_rgb_modified = im.copy()
im_rgb_modified[:, :, 0] = np.clip(im_rgb_modified[:, :, 0] * 1.5, 0, 1)  # 增强红色
im_rgb_modified[:, :, 2] = im_rgb_modified[:, :, 2] * 0.5  # 降低蓝色

# 显示结果
plt.imshow(im_rgb_modified)
plt.axis('off')
plt.show()
````

### 图片的灰度化(RGB三通道取平均)

````python
im_gray = im.mean(axis=2)
plt.imshow(im_gray)
plt.show()
````

### 绘制复合图

`matplotlib`中，还可以使用`subplot()`绘制多个子图将多个图整合在同一张图片中，我们称之为**复合图**

```
plt.subplot(nrows, ncols, index)
```
- `nrows`：子图的行数
- `ncols`：子图的列数
- `index`：当前绘制的是第几个子图（从 1 开始）

````python
x6 = np.arange(0, 3 * np.pi, 0.1)
y7 = np.sin(x6)
y8 = np.cos(x6)

# 前两个参数表示两行一列布局，第三个参数表示先处理第一个图形
plt.subplot(2, 1, 1)

plt.plot(x6, y7)
plt.title('sin')

# 布局不动，处理第二个图形
plt.subplot(2, 1, 2)
plt.plot(x6, y8)
plt.title('cos')

plt.show()
````

## 进阶：搞懂 Axe

<img src="assets/anatomy-of-matplotlib.webp">

在 Matplotlib 中：

> **一个 Figure（图形）可以包含多个 Axes（坐标系）。**
> 每个 Axes 就是一个独立的绘图区，拥有自己的 x 轴、y 轴、标题、图例等。

简单类比：

* **Figure** 像一张画布；
* **Axes** 就是画布上的一个子图区域；
* **Axis**表示具体的坐标轴。


### Axes 的基本创建方式

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()  # 创建一个Figure和一个Axes
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y)
ax.set_title("Sin")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
````


👉 `ax` 就是一个 `matplotlib.axes.Axes` 对象。
所有绘图操作都通过它来完成。


### Axes 的层级结构

Matplotlib 的结构大致是这样的：

```
Figure（画布）
│
├── Axes（坐标系）
│    ├── Axis（坐标轴对象）
│    │     ├── XAxis
│    │     └── YAxis
│    ├── Title / Label / Legend 等文本对象
│    └── Artists（线条、柱状、点、图像等元素）
```

### Axes 常用方法汇总

| 方法                                    | 功能            |
| ------------------------------------- | ------------- |
| `ax.plot(x, y, ...)`                  | 绘制折线图         |
| `ax.scatter(x, y, ...)`               | 绘制散点图         |
| `ax.bar(x, y, ...)`                   | 绘制柱状图         |
| `ax.hist(data, ...)`                  | 绘制直方图         |
| `ax.imshow(img)`                      | 显示图像          |
| `ax.set_title()`                      | 设置标题          |
| `ax.set_xlabel()` / `ax.set_ylabel()` | 设置坐标轴标签       |
| `ax.set_xlim()` / `ax.set_ylim()`     | 设置坐标轴范围       |
| `ax.grid(True)`                       | 显示网格          |
| `ax.legend()`                         | 显示图例          |
| `ax.text(x, y, "text")`               | 在坐标中添加文字      |
| `ax.annotate()`                       | 添加箭头注释        |
| `ax.axhline()` / `ax.axvline()`       | 添加水平/垂直参考线    |
| `ax.spines`                           | 控制边框（上、下、左、右） |


### 多个 Axes 的管理

我们可以使用`plt.subplots(nrows, ncols)`创建多个坐标系。

````python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("sin(x)")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("cos(x)")

axes[1, 0].plot(x, np.exp(x/5))
axes[1, 0].set_title("exp(x/5)")

axes[1, 1].plot(x, np.log(x+1))
axes[1, 1].set_title("log(x+1)")

plt.tight_layout()
plt.show()
````



🧠 小技巧：
`axes` 是一个二维数组（当 nrows、ncols > 1 时），可以用 `axes.flat` 来循环访问。


### 在一个 Figure 中自定义 Axes 位置

有时候你想精准控制子图位置（非网格布局），可以使用：

````python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height]
ax2 = fig.add_axes([0.6, 0.6, 0.25, 0.25])

x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x), 'r')
plt.show()
````


📐 所有数值都是相对于 Figure 的比例（0~1）。
