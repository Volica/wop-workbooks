# Matplotlib

`Matplotlib` ÊòØ `Python` ‰∏≠ÊúÄÂ∏∏Áî®ÁöÑÁªòÂõæÂ∫ì‰πã‰∏ÄÔºåÂπøÊ≥õÁî®‰∫éÊï∞ÊçÆÂèØËßÜÂåñ„ÄÇÂÆÉÊèê‰æõ‰∫Ü‰∏∞ÂØåÁöÑÁªòÂõæÂäüËÉΩÔºåÂèØÁîüÊàêÊäòÁ∫øÂõæ„ÄÅÊü±Áä∂Âõæ„ÄÅÊï£ÁÇπÂõæ„ÄÅÈ•ºÂõæ„ÄÅÁõ¥ÊñπÂõæÁ≠âÂ§öÁßçÂõæÂΩ¢„ÄÇ


`pyplot` ÊòØ `Matplotlib` ÁöÑÊ†∏ÂøÉÊ®°ÂùóÔºåÊèê‰æõÁ±ª‰ºº MATLAB ÁöÑÁªòÂõæÊé•Âè£„ÄÇÂÆÉ‰ª•ÂáΩÊï∞ÂΩ¢ÂºèÊéßÂà∂ÂõæÂΩ¢ÁöÑÂàõÂª∫„ÄÅÊòæÁ§∫ÂíåÊ†∑ÂºèËÆæÁΩÆÔºåÊîØÊåÅÁªòÂà∂ÊäòÁ∫øÂõæ„ÄÅÊü±Áä∂Âõæ„ÄÅÊï£ÁÇπÂõæÁ≠âÂ§öÁßçÂõæË°®„ÄÇÈÄöËøáÁÆÄÂçïÂëΩ‰ª§Âç≥ÂèØÂø´ÈÄüÁîüÊàêÈ´òË¥®ÈáèÂèØËßÜÂåñÁªìÊûúÔºåÂπøÊ≥õÂ∫îÁî®‰∫éÊï∞ÊçÆÂàÜÊûê‰∏éÁßëÁ†îÁªòÂõæ„ÄÇÈÄöÂ∏∏Êù•ËØ¥ÔºåÊàë‰ª¨‰ΩøÁî®ÁÆÄÁü≠ÁöÑÂà´Âêç`plt`Êù•ÂØºÂÖ•ÂÆÉ

## Âü∫Êú¨Êìç‰Ωú

### ÁªòÂà∂ÁÇπÂíåÁ∫ø

````python
x1 = [0, 1, 2, 3, 4]
y1 = [0, 2, 6, 2, 8]

plt.plot(x1, y1, color='blue', linestyle='--', linewidth=2, marker='s')
plt.title('Line Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````

Âú®‰∏äÂõæ‰∏≠ÔºåÊàë‰ª¨Âà©Áî®Âà∞‰∫Ü`plot()`ÂáΩÊï∞Ôºå`plot()`ÂáΩÊï∞ÊòØÁªòÂà∂‰∫åÁª¥ÂõæÂΩ¢ÁöÑÂü∫Á°ÄÔºåÊàë‰ª¨ËøòÂèØ‰ª•Âà©Áî®`scatter`ÂáΩÊï∞Êù•ÁªòÂà∂ÁÇπÔºåÂ¶Ç‰∏ãÂõæÊâÄÁ§∫

````python
x2 = [1, 2, 3, 4, 5]
y2 = [2, 9, 3, 7, 11]

plt.scatter(x2, y2, color='red', marker='o', label='Data Points')
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
````

‰ªé‰∏äÈù¢ÁöÑ‰∏§Âº†ÂõæÈáå‰πüËÉΩÁúãÂá∫ÔºåÂõæ‰∏äÈù¢ÁöÑÁÇπ„ÄÅÁ∫øÊòØÂèØ‰ª•ÈÄöËøáËÆæÁΩÆÂ±ûÊÄßÂèòÂæóÊõ¥Âä†**ÁæéËßÇ**ÁöÑÔºåÊàë‰ª¨Êù•Âàó‰∏Ä‰∏™Ë°®Ê†ºÁÆÄÂçï‰ªãÁªç‰∏Ä‰∏ãÊâÄÊúâÁöÑËøô‰∫õÂ±ûÊÄßÂ∫îËØ•ÊÄé‰πà‰ΩøÁî®üé®

| Â±ûÊÄß                              | ÂèÇÊï∞Á±ªÂûã / Á§∫‰æã                                              | ËØ¥Êòé                           |
| :------------------------------ | :----------------------------------------------------- | :--------------------------- |
| **color / c**                   | `'red'`, `'blue'`, `'g'`, `'#FF8800'`                  | ËÆæÁΩÆÈ¢úËâ≤ÔºåÂèØÁî®È¢úËâ≤Âêç„ÄÅÁÆÄÂÜôÊàñÂçÅÂÖ≠ËøõÂà∂ÂÄº          |
| **linestyle / ls**              | `'-'`, `'--'`, `':'`, `'-.'`, `'none'`                 | ËÆæÁΩÆÁ∫øÂûãÔºàÂÆûÁ∫ø„ÄÅËôöÁ∫ø„ÄÅÁÇπÁ∫ø„ÄÅÁÇπÂàíÁ∫ø„ÄÅÊó†Á∫øÔºâ        |
| **linewidth / lw**              | `1`, `2.5`, `5`                                        | ËÆæÁΩÆÁ∫øÂÆΩ                         |
| **marker**                      | `'o'`, `'s'`, `'^'`, `'x'`, `'+'`, `'d'`, `'*'`, `'.'` | ËÆæÁΩÆÁÇπÁöÑÂΩ¢Áä∂                       |
| **markersize / ms**             | `6`, `10`, `15`                                        | ËÆæÁΩÆÁÇπÁöÑÂ§ßÂ∞è                       |
| **markeredgecolor / mec**       | `'black'`, `'k'`, `'#333333'`                          | ËÆæÁΩÆÁÇπËæπÁºòÈ¢úËâ≤                      |
| **markerfacecolor / mfc**       | `'white'`, `'red'`, `'none'`                           | ËÆæÁΩÆÁÇπÂÜÖÈÉ®Â°´ÂÖÖÈ¢úËâ≤                    |
| **alpha**                       | `0.0 ~ 1.0`                                            | ËÆæÁΩÆÈÄèÊòéÂ∫¶ÔºåÊï∞ÂÄºË∂äÂ∞èË∂äÈÄèÊòé                |
| **label**                       | `'Data1'`, `'sin(x)'`                                  | ËÆæÁΩÆÂõæ‰æãÊñáÂ≠óÔºå‰∏é `plt.legend()` ÈÖçÂêà‰ΩøÁî® |
| **grid()**                      | `plt.grid(True)`                                       | ÊòØÂê¶ÊòæÁ§∫ÁΩëÊ†ºÁ∫ø                      |
| **xlim(), ylim()**              | `plt.xlim(0,10)`                                       | ËÆæÁΩÆÂùêÊ†áËΩ¥ÊòæÁ§∫ËåÉÂõ¥                    |
| **title(), xlabel(), ylabel()** | ÊñáÊú¨Â≠óÁ¨¶‰∏≤                                                  | ËÆæÁΩÆÊ†áÈ¢òÂíåÂùêÊ†áËΩ¥Ê†áÁ≠æ                   |


‰∏ãÈù¢ÁöÑÂõæÂΩ¢ÊòØ‰ΩøÁî®‰∫ÜËØ∏Â§öÂ±ûÊÄßÂêéÁöÑÁ§∫‰æãÔºå‰Ω†‰πüÂèØ‰ª•Âú®Ê≠§Âü∫Á°Ä‰∏äÁªßÁª≠ËøõË°åÁæéËßÇË∞ÉÊï¥üí™

````python
x3 = np.linspace(0, 3, 7)
y3 = 2 ** x3 - 5

plt.plot(
    x3, y3,
    color='purple',        # Á∫øÁöÑÈ¢úËâ≤
    linestyle='--',        # ËôöÁ∫ø
    linewidth=2,           # Á∫øÂÆΩ
    marker='o',            # ÂúÜÂΩ¢Ê†áËÆ∞
    markersize=8,          # Ê†áËÆ∞Â§ßÂ∞è
    markerfacecolor='white',  # Ê†áËÆ∞Â°´ÂÖÖËâ≤
    markeredgecolor='black',  # Ê†áËÆ∞ËæπÊ°ÜËâ≤
    alpha=0.8,             # ÈÄèÊòéÂ∫¶
    label='y = 2x^2-5'
)

plt.title('Line and Marker Style Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

## ÁªòÂà∂ÂáΩÊï∞Âõæ

Âú®`matplotlib`‰∏≠ÔºåÊàë‰ª¨‰πüÂèØ‰ª•ÁªòÂà∂‰∏Ä‰∫õ**ÂáΩÊï∞**ÁöÑÊõ≤Á∫ø

````python
x4 = np.linspace(0, 2 * np.pi, 200)  # Âú® 0 Âà∞ 2œÄ ‰πãÈó¥Âèñ 200 ‰∏™ÁÇπ
y4 = np.sin(x4)

plt.plot(x4, y4, label='y = sinx')
plt.title('sinx')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````

Âú®Âêå‰∏ÄÂº†ÂõæÈáåÔºå‰πüÂèØ‰ª•Êîæ‰∏§Êù°ÂèäÊõ¥Â§öÁöÑÂáΩÊï∞Êõ≤Á∫ø

````python
x5 = np.linspace(0, 2 * np.pi, 200)  # Âú® 0 Âà∞ 2œÄ ‰πãÈó¥Âèñ 200 ‰∏™ÁÇπ
y5 = np.sin(x5)
y6 = np.cos(x5)

plt.plot(x5, y5, label='sin')
plt.plot(x5, y6, linestyle='--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
````

## ÁªòÂà∂Êü±ÂΩ¢Âõæ

* `bar(x, height)`ÔºöÁªòÂà∂ÂûÇÁõ¥Êü±ÂΩ¢Âõæ
* `color`ÔºöÊü±Â≠êÈ¢úËâ≤
* `width`ÔºöÊü±ÂÆΩ
* Ëã•ÈúÄÊ∞¥Âπ≥Â±ïÁ§∫ÔºåÂèØ‰ΩøÁî® `plt.barh()`


````python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 10]

plt.bar(categories, values, color='orange', width=0.6)
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
````

## ÁªòÂà∂È•ºÁä∂Âõæ

* `labels`ÔºöÊØèÈÉ®ÂàÜÊ†áÁ≠æ
* `sizes`ÔºöÊØèÈÉ®ÂàÜÂç†ÊØî
* `autopct`ÔºöÊòæÁ§∫ÁôæÂàÜÊØîÊ†ºÂºè
* `explode`ÔºöÁ™ÅÂá∫ÊüêÈÉ®ÂàÜ
* `startangle`ÔºöÊóãËΩ¨Ëµ∑ÂßãËßíÂ∫¶


````python
labels = ['Apple', 'Banana', 'Cherry', 'Durian']
sizes = [30, 25, 25, 20]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = [0.05, 0, 0, 0]  # Á™ÅÂá∫Á¨¨‰∏Ä‰∏™ÈÉ®ÂàÜ

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=90
)
plt.title('Pie Chart Example')
plt.show()
````

````python
from matplotlib.image import imread
im = imread('assets/lena.png')
plt.imshow(im)
plt.show()
````

### ÂõæÁâáÁöÑË£ÅÂâ™

````python
im_cut = im[50:300, 50:300]
plt.imshow(im_cut)
plt.show()
````

### ÂõæÁâáÁöÑÂèòÊöó

````python
im_darker = im.copy()

# ÊòæÁ§∫ÂõæÂÉè
plt.figure(figsize=(10,6))

for i in range(1,5):
    plt.subplot(2,2,i)
    x = 1 - 0.2*(i-1)
    plt.axis('off') #hide coordinate axes
    plt.title('x={:.1f}'.format(x))
    im_darker = np.clip(im_darker, 0, 1)  # Èò≤Ê≠¢Êï∞ÂÄºË∂ÖÂá∫ËåÉÂõ¥
    plt.imshow(im_darker * x)

plt.show()
````

### ÂõæÁâáÁöÑÈ¢úËâ≤ÊîπÂèò

````python
# ‰øÆÊîπRGBÈ¢úËâ≤
im_rgb_modified = im.copy()
im_rgb_modified[:, :, 0] = np.clip(im_rgb_modified[:, :, 0] * 1.5, 0, 1)  # Â¢ûÂº∫Á∫¢Ëâ≤
im_rgb_modified[:, :, 2] = im_rgb_modified[:, :, 2] * 0.5  # Èôç‰ΩéËìùËâ≤

# ÊòæÁ§∫ÁªìÊûú
plt.imshow(im_rgb_modified)
plt.axis('off')
plt.show()
````

### ÁªòÂà∂Â§çÂêàÂõæ

```
plt.subplot(nrows, ncols, index)
```
- `nrows`ÔºöÂ≠êÂõæÁöÑË°åÊï∞
- `ncols`ÔºöÂ≠êÂõæÁöÑÂàóÊï∞

````python
x6 = np.arange(0, 3 * np.pi, 0.1)
y7 = np.sin(x6)
y8 = np.cos(x6)

# Ââç‰∏§‰∏™ÂèÇÊï∞Ë°®Á§∫‰∏§Ë°å‰∏ÄÂàóÂ∏ÉÂ±ÄÔºåÁ¨¨‰∏â‰∏™ÂèÇÊï∞Ë°®Á§∫ÂÖàÂ§ÑÁêÜÁ¨¨‰∏Ä‰∏™ÂõæÂΩ¢
plt.subplot(2, 1, 1)

plt.plot(x6, y7)
plt.title('sin')

# Â∏ÉÂ±Ä‰∏çÂä®ÔºåÂ§ÑÁêÜÁ¨¨‰∫å‰∏™ÂõæÂΩ¢
plt.subplot(2, 1, 2)
plt.plot(x6, y8)
plt.title('cos')

plt.show()
````

Âú® Matplotlib ‰∏≠Ôºö

> **‰∏Ä‰∏™ FigureÔºàÂõæÂΩ¢ÔºâÂèØ‰ª•ÂåÖÂê´Â§ö‰∏™ AxesÔºàÂùêÊ†áÁ≥ªÔºâ„ÄÇ**

### Axes ÁöÑÂü∫Êú¨ÂàõÂª∫ÊñπÂºè

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()  # ÂàõÂª∫‰∏Ä‰∏™FigureÂíå‰∏Ä‰∏™Axes
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y)
ax.set_title("Sin")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
````


üëâ `ax` Â∞±ÊòØ‰∏Ä‰∏™ `matplotlib.axes.Axes` ÂØπË±°„ÄÇ
ÊâÄÊúâÁªòÂõæÊìç‰ΩúÈÉΩÈÄöËøáÂÆÉÊù•ÂÆåÊàê„ÄÇ


### Axes ÁöÑÂ±ÇÁ∫ßÁªìÊûÑ

Matplotlib ÁöÑÁªìÊûÑÂ§ßËá¥ÊòØËøôÊ†∑ÁöÑÔºö


### Â§ö‰∏™ Axes ÁöÑÁÆ°ÁêÜ

Êàë‰ª¨ÂèØ‰ª•‰ΩøÁî®`plt.subplots(nrows, ncols)`ÂàõÂª∫Â§ö‰∏™ÂùêÊ†áÁ≥ª„ÄÇ

````python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("sin(x)")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("cos(x)")

axes[1, 0].plot(x, np.exp(x/5))
axes[1, 0].set_title("exp(x/5)")

axes[1, 1].plot(x, np.log(x+1))
axes[1, 1].set_title("log(x+1)")

plt.tight_layout()
plt.show()
````

### Âú®‰∏Ä‰∏™ Figure ‰∏≠Ëá™ÂÆö‰πâ Axes ‰ΩçÁΩÆ

ÊúâÊó∂ÂÄô‰Ω†ÊÉ≥Á≤æÂáÜÊéßÂà∂Â≠êÂõæ‰ΩçÁΩÆÔºàÈùûÁΩëÊ†ºÂ∏ÉÂ±ÄÔºâÔºåÂèØ‰ª•‰ΩøÁî®Ôºö

````python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height]
ax2 = fig.add_axes([0.6, 0.6, 0.25, 0.25])

x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x), 'r')
plt.show()
````
